```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = r"/content/drive/MyDrive/Colab Notebooks/GeoPandas/files/ne_110m_admin_0_countries.shp"
world = gpd.read_file(path)
```

Load External GDP File

```python
gdp_path = r"/content/drive/MyDrive/Colab Notebooks/GeoPandas/files/gdp.csv"
gdp_df = pd.read_csv(gdp_path)
```

```python
world_merged = world.merge(gdp_df, on='ADMIN', how='left')

print(world_merged[['ADMIN', 'GDP']].isnull().sum())
```

```python
fig, ax = plt.subplots(figsize=(14, 8))
world_merged.plot(column='GDP', cmap='YlGnBu', legend=True, edgecolor='black', ax=ax)
plt.title("Countries by GDP")
plt.axis("off")
plt.show()
```
