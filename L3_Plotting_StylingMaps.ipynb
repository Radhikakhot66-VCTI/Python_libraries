```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
path = r"/content/drive/MyDrive/Colab Notebooks/GeoPandas/files/ne_110m_admin_0_countries.shp"
world = gpd.read_file(path)
```

```python
world.plot(figsize=(12, 6), edgecolor='black')
plt.title(" World Map - Default Style")
plt.show()
```

Color countries by continent

```python
world.plot(column="CONTINENT", cmap="Set3", figsize=(12, 6), legend=True, edgecolor="black")
plt.title("Countries Colored by Continent")
plt.show()
```

Highlight India

```python
india = world[world["ADMIN"] == "India"]

fig, ax = plt.subplots(figsize=(10, 8))
world.plot(ax=ax, color='lightgray', edgecolor='white')
india.plot(ax=ax, color='green', edgecolor='black')
ax.set_title("India Highlighted on World Map")
plt.axis("off")
plt.show()
```
