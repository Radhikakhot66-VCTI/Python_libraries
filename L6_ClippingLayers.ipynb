```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load country polygons
country_path = r"/content/drive/MyDrive/Colab Notebooks/GeoPandas/files/ne_110m_admin_0_countries.shp"
world = gpd.read_file(country_path)

# Load rivers shapefile
rivers_path = r"/content/drive/MyDrive/Colab Notebooks/GeoPandas/files/riverFiles/ne_110m_rivers_lake_centerlines.shp"
rivers = gpd.read_file(rivers_path)
```

Filter for India

```python
india = world[world["ADMIN"] == "India"]
rivers = rivers.to_crs(india.crs)

```

Clip Rivers to India's Boundary

```python
rivers_in_india = gpd.clip(rivers, india)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
india.plot(ax=ax, color='whitesmoke', edgecolor='black')
rivers_in_india.plot(ax=ax, color='blue', linewidth=1)
plt.title("Rivers Within India")
plt.axis("off")
plt.show()
```
