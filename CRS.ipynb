```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
path = r"/content/drive/MyDrive/Colab Notebooks/GeoPandas/files/ne_110m_admin_0_countries.shp"
world = gpd.read_file(path)
```

Print current crs


```python
print("Current CRS:", world.crs)
```

```python
india = world[world['ADMIN'] == 'India']
india.plot(color='lightblue', edgecolor='black')
plt.title("India in EPSG:4326 (lat/lon)")
plt.show()
```

reprojecting using .to_crs

```python
india_3857 = india.to_crs("EPSG:3857")
```

```python
india_3857.plot(color='orange', edgecolor='black')
plt.title("India in EPSG:3857 (meters)")
plt.show()
```

Compare area before and afte

```python
print("Area in EPSG:4326:", india.geometry.area.values[0])       # degrees² (not meaningful)
print("Area in EPSG:3857:", india_3857.geometry.area.values[0]) # meters² (real-world area)
```
