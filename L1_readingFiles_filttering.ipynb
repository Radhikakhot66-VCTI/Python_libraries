```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
path = r"/content/drive/MyDrive/Colab Notebooks/GeoPandas/files/ne_110m_admin_0_countries.shp"
world = gpd.read_file(path)
```

```python
print(world.columns)            # List all columns
print(world.head(2))            # Preview first rows
print(world.crs)                # View Coordinate Reference System
```

Plotting full world

```python
world.plot(figsize=(12, 6), edgecolor='black')
plt.title("World Countries Map")
plt.show()
```

Filter for a specific country (India)

```python
india = world[world["ADMIN"] == "India"]
```

```python
india.plot(color='lightgreen', edgecolor='black')
plt.title("India Boundary")
plt.show()
```
